## **User-defined functions** <a class="tocSkip">

```python
# 1. Preview a data frame
def preview(df):
    """Receives data frame.
    Returns a summary of attributes including dtype, proportion of missing values,
    number of unique values, and sample."""
    
    dfCheck= []
    for i in df.columns:
        dfCheck.append([
                    i,
                    df[i].dtype,
                    df[i].isna().sum(),
                    round(df[i].isna().sum() / len(df) * 100, 2),
                    df[i].nunique(),
                    df[i].drop_duplicates().values])
        
    return pd.DataFrame(dfCheck, columns = ['Features', 'DataType', 'Null', 'Proportion', 'Uniques', 'UniqueSample'])

# 2. Outlier
def outlier (df, numerical):
    """Receives data frame and numerical columns.
    Returns a list of all records of outliers index."""
    
    q1 = df[numerical].quantile(0.25)
    q2 = df[numerical].quantile(0.5)
    q3 = df[numerical].quantile(0.75)

    iqr = q3 - q1
    o = df[(df[numerical] < (q1 - (1.5 * iqr))) | (df[numerical] > (q3 + (1.5 * iqr)))].index
    
    return df.loc[o]

# 3. Substract days
def subtract_days_from_date(date, days):
    """Receives a date and subtract days from a date
    then return the date."""
    
    subtracted_date = pd.to_datetime(date) - dt.timedelta(days=days)
    subtracted_date = subtracted_date.strftime("%Y-%m-%d")

    return subtracted_date

# 4. Chi-squared post-hoc test
def get_asterisks_for_pval(p_val):
    """Receives the p-value and returns asterisks string."""
    
    if p_val > 0.05:
        p_text = "ns"  # Above threshold => not significant
    elif p_val < 1e-4:  
        p_text = '****'
    elif p_val < 1e-3:
        p_text = '***'
    elif p_val < 1e-2:
        p_text = '**'
    else:
        p_text = '*'
    
    return p_text

def chisq_and_posthoc_corrected(df):
    """Receives a data frame and performs chi2 test and then post hoc.
    Prints the p-values and corrected p-values (after FDR correction)"""
    
    # Start by running chi2 test on the matrix
    chi2, p, dof, ex = chi2_contingency(df, correction = True)
    print(f"Chi2 result of the contingency table: {chi2}, p-value: {p}")
    
    # Post-hoc
    all_combinations = list(combinations(df.index, 2))  # gathering all combinations for post-hoc chi2
    p_vals = []
    print("Significance results:")
    for comb in all_combinations:
        new_df = df[(df.index == comb[0]) | (df.index == comb[1])]
        chi2, p, dof, ex = chi2_contingency(new_df, correction=True)
        p_vals.append(p)
        # print(f"For {comb}: {p}")  # uncorrected

    # Checking significance
    # Correction for multiple testing
    reject_list, corrected_p_vals = multipletests(p_vals, method='fdr_bh')[:2]
    for p_val, corr_p_val, reject, comb in zip(p_vals, corrected_p_vals, reject_list, all_combinations):
        print(f"{comb}: p_value: {p_val:5f}; corrected: {corr_p_val:5f} ({get_asterisks_for_pval(p_val)}) reject: {reject}")
```

***
* $TP$: True positive <br>
Description: Potential churners that are correctly predicted as churners <br>

* $FP$: False positive <br>
Description: Potential non-churners that are incorrectly predicted as churners <br>
Consequence: **Relatively small** <br>
Comment: Waste of marketing expenses including time and budget for targeting the already loyal customers

* $TN$: True negative <br>
Description: Potential non-churners that are correctly predicted as non-churners <br>

* $FN$: False negative <br>
Description: Potential churners that are incorrectly predicted as non-churners <br>
Consequence: **Very high** <br> 
Comment: Even if the company does not exert any marketing expense to target the potential churners, the opportunity cost (loss of potential lifetime value) of this misclassification is very high
***
Based on the consequences, we want to mainly:
1. Maximize $TP$ <br>
1. Minimize $FN$ <br>
1. Aim for a satisfactory $FP$ after the first 2 conditions have been satisfied

This approach would give us a model that could prioritize correctly predicting the potential churners while also avoid misclassifiying potential churners as non-churners. Furthermore, given the imbalanced nature of the dataset (shown in the next section), we chose **precision-recall curve (PR curve)** as the evaluation metric for this problem. PR curve is used when the problem of class imbalance is associated to the presence of a low sample size of minority instances (in this case is the rarity of churners), as the estimates can be unreliable [Fernndez et al., 2018; Hanczar et al., 2010]. Specifically, we used the metric **average precision** (`AP`) as it is the single metric that can summarize PR curve. AP is given by the formula [Zhu, 2004]:

$$\text{Average precision} = \sum_{k=1}^NP(k)\Delta R(k)$$
Where:
- $P(k) = $ The precision at cutoff of $k$ threshold
- $\Delta R(k) = $ The change in recall that happened between cutoff $k-1$ and $k$ thresholds
- $N =$ Total number of data points

This reasoning is based on the underlying business problem. Indeed, The e-commerce companys main motivation is to detect the maximum number of churners that is having a high value of $TP$. On the other hand, it is much more acceptable to classify non-churners as churners $(FP)$, given that the action taken towards expected churners only would result in an extra marketing expense. Therefore, having a low `precision` while maximizing `recall`, even if not optimal, is much more important than aiming to satisfy both measures equally. However, before taking such measure, it is preferred to prioritize a model that produces a high `AP` first since it represents both high `recall` and high `precision` (i.e. the best model for the business case). Afterwards, the focus would be put in choosing a high `recall` score and then aim for a satisfactory `precision`, all of which could be achieved by using PR-curve & `AP` as the evaluation metric.

At a glance, there seem to be differences in proportions of churn between groups. However, to confirm this, **test of homogeneity** should be performed.

A test of homogeneity tests the null hypothesis that different populations have the same proportions of some characteristics.

#### **Variable: `PreferredOrderCat`**

**Hypothesis** <br> <br>
$H_0:$ $p_0 = p_1 = \ldots = p_k$ <br>
$H_1:$ $\exists \, i,j:p_i \neq p_j$ <br>
$\alpha = 0.05$

```python
df_temp[['PreferredOrderCat', 'Churn']]
q2_PreferredOrderCat = pd.crosstab(df_temp['PreferredOrderCat'], df_temp['Churn'])
q2_PreferredOrderCat
```

```python
tstat, pval, dof, e = chi2_contingency(q2_PreferredOrderCat.values.tolist())
print(tstat, pval)

# Conclusion
if pval < 0.05 :
    print('Reject H0')
else: 
    print('Accept H0')
```

There is not enough evidence (at $\alpha = 5\%$) to say that the `PreferredOrderCat` groups have the same proportions of customers that churn and do not churn. The slight difference might be caused due to chance alone.

$H_0:$ $p_0 = p_1$ <br>
$H_1:$ $p_0 \neq p_1$

```python
df_temp[['Complain', 'Churn']]
q2_Complain = pd.crosstab(df_temp['Complain'], df_temp['Churn'], margins = True)
q2_Complain
```

```python
x0 = q2_Complain.loc[0, 1]
n0 = q2_Complain.loc[0,'All']
p0 = x0/n0
print('Churn proportion of customers that do not complain', p0)

x1 = q2_Complain.loc[1, 1]
n1 = q2_Complain.loc[1,'All']
p1 = x1/n1
print('Churn proportion of customers that do complain', p1)

# Success
success = np.array([x0, x1])

# Total number of observation
obs = np.array([n0, n1])

tstat, pval = proportions_ztest(success, obs)
print(tstat, pval)

# Conclusion
if pval < 0.05 :
    print('Reject H0')
else: 
    print('Accept H0')
```

There is not enough evidence (at $\alpha = 5\%$) to say that the `Complain` groups have the same proportions of customers that churn and do not churn.

**REMARKS**
> - Customers that complain have significantly different churn proportion than customers that do not complain
> - From the samples in this data set, $28\%$ of customers that complain ended up churning while only $9\%$ of customers that do not complain ended up churning
> - The company should consider to improve its customer service to reduce the decrease the amount of complaints filed by the customers

## **Benchmarking**

### **Baseline model**

First, we want to evaluate a baseline model that predicts a random class label in proportion to the base rate of each class. To achieve this, we use `DummyClassifier` **that makes predictions that ignore the input features** with parameter `strategy = 'stratified'`. This classifier serves as a simple baseline to compare against other more complex classifiers and should result in a average precision (`AP`) (Since we are using PR-curve as the evaluation metric).

```python
dummy.fit(x_train, y_train)
positive_proba = dummy.predict_proba(x_test)[:, 1] # Retrieve probability just for the positive class
AP = average_precision_score(y_test, positive_proba)
print(f'Dummy classifier AP: {AP:.4f}')
```

- The dummy classifier `AP` is $0.1592\,\,(15.92\%)$, which is very close to our proportion of the positive class (Churned customers) at $17.10\%$ (This small error might be caused by the removal of missing values in the pipeline)
- This is just as described by Saito & Rehmsmeier, where unlike `ROC AUC` whose baseline is always at $0.5$, `AP` baseline has the expected value close to the proportion of positive instances in the dataset $\left(\frac{\text{Actual positive class}}{\text{Total class}}\right)$  [Saito & Rehmsmeier, 2015]

**REMARKS**
> Therefore, all other model performance (`AP`) should at least bigger than $0.1592$ to be considered skillful.

### **Cross-validation**

Now, we evaluate candidate models using repeated stratied k-fold cross-validation. 
- The k-fold cross-validation procedure provides a good general estimate of model performance that is not too optimistically biased, at least compared to a single train-test split
- The parameter `n_splits = 10` is used, meaning each fold will contain about $\frac{3265}{10}$ or about $326$ examples (Ignoring missing values)
- Since it's been established that the model would optimize PR curve to solve the business problem, `scoring = 'average_precision'` is used
- Stratied means that each fold will contain the same proportion of examples by class, that is $1 : 4.8 \,\, (\text{Churn} : \text{No churn})$

```python
models = [dummy, logreg, tree, rf, tree_ada, xgb, lgb]
method = ['Dummy classifier', 'Logistic regression', 'Decision tree', 'Random forest', 'Adaboost (Tree)', 'XGBoost', 'LightGBM']
score_cv = []
mean = []
std = []

for i in models:
    estimator = Pipeline(
                            [
                                ('transformer', transformer),
                                ('model', i)
                            ]
                        )
    model_cv = cross_val_score(estimator, x_train, y_train, cv = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 2022), scoring = 'average_precision')
    score_cv.append(model_cv.round(4))
    mean.append(model_cv.mean().round(4))
    std.append(model_cv.std().round(4))
    
cv_summary = pd.DataFrame(
                            {
                             'Method': method,
                             'CV score': score_cv,
                             'Mean AP': mean,
                             'Std. AP': std
                            }
                         )
cv_summary
```

The author chose to explore oversampling methods instead of increasing complexity:

```python
random_oversampler = RandomOverSampler(sampling_strategy = 'minority', random_state = 2022)
smote = SMOTE(sampling_strategy = 'minority', random_state = 2022)
smotetomek = SMOTETomek(sampling_strategy = 'minority', random_state = 2022)
```

### **Cross-validation with imbalanced learning**

```python
# Chosen model from benchmarking
models = xgb
method = 'XGBoost'

# Imbalanced learning: Sampling methods
imb_models = [random_oversampler, smote, smotetomek]
imb_method = ['XGBoost + No sampling','XGBoost + Random oversampler', 'XGBoost + SMOTE', 'XGBoost + SMOTE-Tomek']

# Cross-validation metrics
scoring = ['precision', 'recall', 'average_precision']
mean_precision = []
std_precision = []
mean_recall = []
std_recall = []
mean_ap = []
std_ap = []

# XGBoost only (No sampling)
estimator = Pipeline(
                        [
                            ('transformer', transformer),
                            ('model', models)
                        ]
                    )
for j in scoring:
    model_cv = cross_val_score(estimator, x_train, y_train, cv = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 2022), scoring = j)
    if j == 'precision':
        mean_precision.append(model_cv.mean().round(4))
        std_precision.append(model_cv.std().round(4))
    elif j == 'recall':
        mean_recall.append(model_cv.mean().round(4))
        std_recall.append(model_cv.std().round(4))
    else:
        mean_ap.append(model_cv.mean().round(4))
        std_ap.append(model_cv.std().round(4))

# XGBoost + sampling methods
for i in imb_models:
    estimator = Pipeline(
                            [
                                ('transformer', transformer),
                                ('sampling', i),
                                ('model', models)
                            ]
                        )
    for j in scoring:
        model_cv = cross_val_score(estimator, x_train, y_train, cv = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 2022), scoring = j)
        if j == 'precision':
            mean_precision.append(model_cv.mean().round(4))
            std_precision.append(model_cv.std().round(4))
        elif j == 'recall':
            mean_recall.append(model_cv.mean().round(4))
            std_recall.append(model_cv.std().round(4))
        else:
            mean_ap.append(model_cv.mean().round(4))
            std_ap.append(model_cv.std().round(4))
    
cv_summary = pd.DataFrame(
                            {
                             'Method': imb_method,
                             'Mean_precision': mean_precision,
                             'Std._precision': std_precision,
                             'Mean_recall': mean_recall,
                             'Std._recall': std_recall,
                             'Mean_AP': mean_ap,
                             'Std._AP': std_ap
                            }
                         )

cols = ['Mean_precision', 'Std._precision']
cv_summary['Precision'] = cv_summary[cols].apply(lambda row: '  '.join(row.values.astype(str)), axis = 1)
cols = ['Mean_recall', 'Std._recall']
cv_summary['Recall'] = cv_summary[cols].apply(lambda row: '  '.join(row.values.astype(str)), axis = 1)
cols = ['Mean_AP', 'Std._AP']
cv_summary['Average precision'] = cv_summary[cols].apply(lambda row: '  '.join(row.values.astype(str)), axis = 1)

cv_summary[['Method', 'Precision', 'Recall', 'Average precision']]
```

## **Hyperparameter tuning**

Now that we know the default performance of the model, we can take the next step and improve the models generalization performance by tuning its hyperparameters. Specifically, the approach that is taken is `GridSearchCV` which involves using brute force to find the best combination of hyperparameters in a defined range. It is the most computationally exhaustive approach but also more often exhibits the best result. 

Furthermore, using `StratifiedKFold` in combination with grid search is a useful approach for fine-tuning the performance of a machine learning model. In nested cross-validation, we have an outer `StratifiedKFold` loop to split the data into training and test folds, and an inner loop is used to select the model using `StratifiedKFold` on the training fold. After model selection, the test fold is then used to evaluate the model performance [Rachska & Mirjalili, 2019]. In this particular study, a number of fold of 5 (`n_splits` = `5`) was chosen as a balance of computing power and robustness for the tuning process.

### **Default hyperparameters**

## **PR-curve & threshold moving**

Now that we've acquired the best tuned model for `AP` (Indicating the best performance for an imbalanced classification problem), we can further improve the model by moving the probability threshold. The F-measure would be used to move the probability threshold such that the model could achieve a highest `recall` score then aim for satisfactory `precision` score. First, several $\beta$ values would be tested for the F-measure. Then, we find the threshold that holds the highest value for that specific F-measure.

$$F_\beta = (1+\beta^2)\frac{\text{Precision}\cdot\text{Recall}}{(\beta^2\cdot\text{Precision}) + \text{Recall}} \,\, \begin{cases} \beta < 1 \Rightarrow \text{Optimize for precision} \\ \beta = 1 \Rightarrow \text{Optimize for both precision and recall} \\ \beta > 1 \Rightarrow \text{Optimize for recall}\end{cases}$$

Here, 3 $\beta$ values are chosen to optimize for `recall`, starting from baseline value $(1)$ to larger values to see the trade-off between `precision` and `recall`
1. $\beta = 1 \implies$ Baseline value, pose a balance between `precision` and `recall`
1. $\beta = 1.6 \implies$ Preemptively chosen as this specific number would satisfy the business problem of maximizing $TP$ and minimizing $FN$ then aim for satisfying $FP$
1. $\beta = 2 \implies$ Focuses solely on maximizing `recall`

```python
positive_proba = tuned_estimator.predict_proba(x_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, positive_proba)

f1score = (2 * precision * recall) / ((precision) + recall)
ix1 = np.argmax(f1score)
print('Best Threshold = %f, Precision = %.4f, Recall = %.4f, F_1 = %.4f' % (thresholds[ix1], precision[ix1], recall[ix1], f1score[ix1]))

f1_6score = ((1 + 1.6**2) * precision * recall) / ((1.6**2 * precision) + recall)
ix2 = np.argmax(f1_6score)
print('Best Threshold = %f, Precision = %.4f, Recall = %.4f, F_1.6 = %.4f' % (thresholds[ix2], precision[ix2], recall[ix2], f1_6score[ix2]))

f2score = (5 * precision * recall) / ((4 * precision) + recall)
ix3 = np.argmax(f2score)
print('Best Threshold = %f, Precision = %.4f, Recall = %.4f, F_2 = %.4f' % (thresholds[ix3], precision[ix3], recall[ix3], f2score[ix3]))
```

    **Assumptions**
    - Average customer lifetime value (CLV) (US) = $\$340$ (Rounded, assuming they purchase apparel, footwear, accesories) [https://www.statista.com/study/42335/ecommerce-report/]
    - Average marketing expense per customer (US) = $\$66$ (Rounded, assuming they come from retail industry) [https://www.demandjump.com/blog/customer-acquisition-cost-by-industry]
    - Let there be $150$ customers at a given time with an 1:5 proportion of churn and not churn (Similar to imbalanced proportion in this data):
        - $25$ churners $(POS)$
        - $125$ non-churners $(NEG)$
    ***
    **Without model**
    - All 150 customers are given marketing promotion regardless of their potential to churn
    - Total marketing expense: $150 \times \$66 = \$9,900$
    - Total customer lifetime value from potential churners but prevented by marketing promotion: $25 \times \$340 = \$8,500$
    - Total wasted expenses given to customers that do not churn: $125 \times \$66 = \$8,250$
    - Total save: $\$8,500 - \$8,250 = \$250$
    ***
    **With model**
    - $86\%$ of $25$ $(22)$ churners that are given marketing promotion $\implies TP = 22$
    - $11\% \,\, [1-\text{Recall(-)}]$ of $125$ $(14)$ non-churners that are given marketing promotion $\implies FP = 14$
    - $TN + FP = NEG \implies TN = 111$
    - $TP + FN = POS \implies FN = 3$
    - Summary: $TP = 22, \,\, FP = 14, \,\, TN = 11, \,\, FN = 3$
    - Marketing promotion is only given to customers that are predicted to churn i.e. $TP + FP$
    - Total marketing expense: $TP + FP \implies (22 + 14) \times \$66 = \$2,376$
    - Total customer lifetime value from potential churners but prevented by marketing promotion: $22 \times \$340 = \$7,480$
    - Total wasted expenses given to non-churners: $14 \times \$66 = \$924$
    - Total loss of customer lifetime value from churners that are not given marketing promition: $3 \times \$340= \$1,020$
    - Total save: $\$7,480 - \$924 - \$1,020 = \$5,536$
    ***
    **Bottomline**
    - Applying model would result in over $\sim 22$ times greater cost saving than without using model
